```python
server_logs = [
    {"service": "nginx", "status": "active", "cpu_usage": 99},
    {"service": "postgresql", "status": "active", "cpu_usage": 45},
    {"service": "redis", "status": "inactive", "cpu_usage": 0},
    {"service": "docker", "status": "active", "cpu_usage": 22},
]
print(f"Loaded {len(server_logs)} server logs into memory.")
```

```python
high_load_servers = [
    s["service"].upper() for s in server_logs
    if s["status"] == "active" and s["cpu_usage"] > 20
]
print(f"Alert! High load on: {high_load_servers}")
```

```python
import httpx
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("NEWS_API_KEY")

async with httpx.AsyncClient() as client:
    url= f"https://newsapi.org/v2/top-headlines?country=us&apiKey={api_key}"
    response= await client.get(url)
    response.raise_for_status()
    news_data = response.json()

titles= [a["title"] for a in news_data.get("articles", [])[:10]]
for t in titles:
    print(f"Title: {t}")
```
