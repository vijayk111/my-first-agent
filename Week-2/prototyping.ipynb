```python
import time
print("Calling AI...")
time.sleep(2)
ai_dummy_response= {"text": "The Ubuntu server is running smooth.", "status": "200"}
print("Response stored in memory!")
```

```python
from pydantic import BaseModel

class StatusReport(BaseModel):
    message: str
    code: int

clean_data = StatusReport(
    message=ai_dummy_response["text"],
    code= int(ai_dummy_response["status"])
)

print(clean_data)
```
