```python
import pandas as pd

data = {
    'timestamp': ['2024-03-01 10:00', '2024-03-01 10:05', '2024-03-01 10:10', '2024-03-01 10:15', '2024-03-01 10:20'],
    'server_id': ['web-01', 'db-01', 'web-01', 'db-01', 'web-01'],
    'cpu_usage': [15.5, 88.2, 18.1, 92.5, 12.0],
    'memory_usage': [40, 75, 42, 78, 38],
    'status': ['OK', 'CRITICAL', 'OK', 'CRITICAL', 'OK']
}

df_initial = pd.DataFrame(data)
df_initial.to_csv('server_metrics.csv', index=False)
print("CSV created successfully!")
```

```python
df = pd.read_csv('server_metrics.csv')

print("---First 3 Rows---")
print(df.head(3))

print("---Data Info---")
print(df.info())

critical_events = df[df['status'] == 'CRITICAL']

avg_cpu = df.groupby('server_id')['cpu_usage'].mean()

print("\n---Critical Events ----")
print(critical_events)

print("\n---Average CPU Usage per server----")
print(avg_cpu)
```
